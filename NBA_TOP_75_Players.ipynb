```python
import pandas as pd
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/NBA_75th_Anniversary_Team'
```

```python
ls_tables = pd.read_html(url)
df = ls_tables[1]

df.head()
```

```python
df.rename(
    columns = {
    "Name[5]": "Name",
    "Team(s) played for (years)[note 3]": "Teams_Years",
    "Championships won": "Championships_won",
    "MVP won": "MVP_won",
    "Finals MVP won": "Finals_MVP_won"
    },
    inplace = True
)

df.head()
```

```python
df.drop(
    labels = ['Ref.'],
    axis = 1,
    inplace = True
)

df.head()
```

```python
df["Championships_won"].map(lambda x: x if x is np.nan else [int(item.strip()) for item in x.strip('()').split(',')])
```

```python
def clean_dates(list_of_dates: str) -> list:
    """list_of_dates: str -> A string of the format: '(1000, 2000, 3000)'
    
    Takes in a string of comma separated dates and returns a list of integer dates."""
    
    if list_of_dates is np.nan:
        return list_of_dates
    
    else:
        ls = list_of_dates.strip('()').split(',') # returns a list of string values of dates
        ls_int = [int(item.strip()) for item in ls] # Converts each element of the list into integer. Removes white spaces

        return ls_int
    
df["Championships_won"] = df["Championships_won"].map(clean_dates)
df["MVP_won"] = df["MVP_won"].map(clean_dates)
df["Finals_MVP_won"] = df["Finals_MVP_won"].map(clean_dates)

df.head()
```
